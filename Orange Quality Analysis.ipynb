# 1.Is there a correlation between the size (Size (cm)) and the weight (Weight (g)) of the fruit?

```python
data=pd.read_csv(r"Orange Quality Data.csv",usecols=['Size (cm)','Weight (g)'])
```

```python
total=data.groupby('Size (cm)').corr()
total.sort_values(by='Size (cm)',ascending=False)
```

- there is a positive correlation between sixe and weight 

# 2.How does the sweetness (Brix) of the fruit correlate with its acidity (pH)

```python
data=pd.read_csv(r"Orange Quality Data.csv",usecols=['Brix (Sweetness)','pH (Acidity)'])
```

```python
sns.scatterplot(x='Brix (Sweetness)',y='pH (Acidity)',data=data)
```

- using the scatterplot , we can see there is no correlation between the Brix(Sweetness) and pH (Acidity)
- In the above scatterplot;
- Points are scattered randomly : No Correlation
- Positive correlation: Points trend upwards.
- Negative correlation: Points trend downwards

# 3.How does the softness (Softness (1-5)) of the fruit relate to its ripeness (Ripeness (1-5))

```python
data=pd.read_csv(r"Orange Quality Data.csv")
```

# 4.Do fruits with blemishes (Blemishes (Y/N)) tend to have lower ripeness or softness levels?

```python
data=pd.read_csv(r"Orange Quality Data.csv",usecols=['Softness (1-5)','Ripeness (1-5)','Blemishes (Y/N)'])
```

```python
data.head(1)
```

```python
total=data.groupby('Blemishes (Y/N)').sum()
total.sort_index(ascending=True)
```

- Blemishes (N) has very high softness and Ripeness
- Blemishes with Minor(N), Split Skin(N), Bruise(Y) and Sun burn(Y) has less Softness and less Ripeness
- and Blemishes with Bruising(Y), Minor insect damage(Y), Minor(Y), Mold spot(Y), Scars(Y), Split skin(Y), Sunburn Patch(Y) has High Softness and High Ripeness

 # 5.How do different fruit varieties (Variety) differ in terms of size, weight, sweetness, acidity, and ripeness?

```python
data=pd.read_csv(r"Orange Quality Data.csv", usecols=['Size (cm)','Weight (g)','Brix (Sweetness)','pH (Acidity)','Ripeness (1-5)','Variety'])
```

```python
data.head(1)
```

```python
total=data.groupby('Variety').sum()
total.sort_index(ascending=True)
```

- Size and Weight: Cara Cara and Star Ruby are heavier and larger, and smaller varieties like Tangelo and Blood Orange are more small and lighter.
- Sweetness (Brix): Temple and Cara Cara are the sweetest, Tangelo (Hybrid) and Blood Orange are much less sweet.
- Acidity (pH): Blood Orange and Tangelo have high acidity (lower values mean high acidity)
- Ripeness: Cara Cara and Temple are fully ripe.
